```python
import logging
import threading
logging.basicConfig(level=logging.DEBUG,
                   format='[%(levelname)s] (%(threadName)-10s) %(message)s',
                    )

def experiment(blocks=7):
    logging.debug('Starting with {0} blocks'.format(blocks))
    for block in range(0, blocks):
        logging.debug('** Block ' + str(block))
        socket.send('record on True')
        message = socket.recv()
        logging.debug("  Received reply ", "[", message, "]")
        
        for request in range (0, 150):
            wave_file = wave_files[np.random.randint(0, len(wave_files))]
            time.sleep(np.random.randint(5, 10)) 
            cmd_line = 'trial type passive number %i stim_file %s' % (request, wave_file)
            logging.debug(cmd_line)
            socket.send(cmd_line)
            #  Get the reply.
            message = socket.recv()
            logging.debug("  Received reply ", request, "[", message, "]")
    
        logging.debug('Blocked finished, stopping recording')
        socket.send('record on False')
        message = socket.recv()
        logging.debug("  Received reply ", request, "[", message, "]")
        time.sleep(1800)
    
t = threading.Thread(name='experiment', target=experiment)
t.start()
```
